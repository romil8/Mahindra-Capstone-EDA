```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Area / Locality'] = df['Area / Locality'].replace({'.':'NA'})
df['Area / Locality'] = df['Area / Locality'].replace({',':'NA'})
df['Area / Locality'].value_counts()
```

```python
corr = df.corr()
g = sns.heatmap(corr,  vmax=.7, center=0, linewidths=1.2, annot=True,fmt='.1f', cmap='coolwarm')
sns.despine()
g.figure.set_size_inches(14,10)
    
plt.show()
```

```python
sns.set(style="ticks")

filtered_df = df[(df['Area / Locality'].isin(['NA','NASHIK','THANE','NAGPUR','CHENNAI','FARIDABAD','VAPI','PUNE','KLP','NELLORE','MOGA'])) & 
                      (df['Make'].isin(['GENERAL MOTORS', 'TATA MOTORS', 'MARUTI SUZUKI', 'HYUNDAI',
       'TOYOTA', 'HONDA', 'FIAT', 'FORD', 'SKODA', 'MAHINDRA &  MAHINDRA',
       'VOLKSWAGEN', 'NISSAN', 'RENAULT', 'MERCEDES BENZ',
       'MITSUBISHI MOTORS', 'DAEWOO', 'BMW', 'FORCE', 'SONALIKA', 'VOLVO',
       'HINDUSTAN MOTORS', 'AUDI', 'PREMIER\xa0AUTOMOBILES', 'PORCHE',
       'LAND ROVER', 'JEEP', 'MORRIS', 'SAN MOTORS']))]

df_plot = filtered_df.groupby(['Area / Locality', 'Make']).size().reset_index().pivot(columns='Area / Locality', index='Make', values=0).reset_index()

g = df_plot.set_index('Make').T.plot(kind='bar', stacked=True, color=sns.color_palette())
sns.despine()
g.figure.set_size_inches(20,12)
    
plt.show()
```
